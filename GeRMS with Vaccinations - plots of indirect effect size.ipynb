## Plot 2: Coverage vs. effect size

```python
# coverage
cov = np.array([1., 2./3, .5, 1./3, 1./5, 1./10, 1./30, 1./100])
# number of clusters needed for indirect and direct measurements at 25% efficacy
n_i25 = [31.36, 172.329, 393.36, 1089.95, 3500.45, 15459.8, 148052., 1.68*10**6]
n_d25 = [35.0638, 109.503, 187.299, 328.643, 537.873, 811.835, 1100.14, 1232.41]
# number of clusters needed for indirect and direct measurements at 50% efficacy
n_i50 = [31.36, 31.36, 35.9881, 160.07, 636.963, 3160.87, 32274.1, 373784.]
n_d50 = [31.36, 31.36, 33.6032, 69.3774, 117.254, 172.776, 224.622, 246.61]
# number of clusters needed for indirect and direct measurements at 75% efficacy
n_i75 = [31.36, 31.36, 31.36, 35.6492, 202.009, 1146.01, 12461.4, 147057.]
n_d75 = [31.36, 31.36, 31.36, 32.387, 49.125, 65.0556, 77.6912, 82.5551]
```

```python
fig, ax1 = plt.subplots(figsize = (8,6))
ax1.plot(100*cov, marker = 'o', color = 'k')
ax1.set_ylabel("Coverage (%)", fontsize = 14)
ax1.set_xlabel('Cluster Size', fontsize = 14)
ax1.set_xticklabels(["0", "100", "150", "200", "300", "500", "1000", "3000", "10000"], \
                    fontsize = 14)
ax1.tick_params(axis = "both", labelsize = 14)
ax2 = ax1.twinx()
ax2.plot(n_i25, marker = 'o', color = 'r')
ax2.plot(n_d25, marker = 'o', color = 'g')
ax2.set_ylabel("Required Number of Clusters", fontsize = 14)
ax2.set_yticks([25,50,75,100,125,150,175,200,225,250])
ax2.set_ylim(0,250)
ax2.tick_params(axis = "both", labelsize = 14)
ax1.set_title("25% vaccine efficacy\n100 vaccinated per cluster", fontsize = 20)
ax2.legend(["Indirect Effect", "Direct Effect"], \
           fontsize=14, loc = "center right");
plt.savefig("/Users/dtcitron/Documents/MASH/Bioko_Macro/GeRMS_effect_size_plots/n_clusters_25vaxeff.pdf")
```
